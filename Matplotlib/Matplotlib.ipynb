If we want to plot within an IPython notebook we can use %matplotlib. It comes with two different commands

We will generally use %matplotlib inline

```python
import numpy as np
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
%matplotlib inline
```

Generating Sin and Cos Graph using Matplotlib

```python
x=np.linspace(0,4*np.pi,100)
sin_y=np.sin(x)
cos_y=np.cos(x)
plt.plot(x,sin_y,label='sine')
plt.plot(x,cos_y,label='cos')
plt.legend(loc='upper right')
plt.show()
```

Legend is used to add labels to the graph given in the plot function. The loc argument takes an integer or string that specifies the location of the legend on the plot. different values for loc, such as lower right, center, upper right, lower left, etc.
title function adds a title to the plot
'-' will construct a line like in above plots

```python
plt.plot(x,sin_y,label='*')
plt.plot(x,cos_y,label='--')
plt.title('Sine and Cos wave')
```

In an IPython notebook the graph will be plotted even without using the plt.show() but in some conditions when we are working on big datasets and want to find some inference we use the plt.show() so it is generally a good practice to use it and also if we are writing a python script then also it's necessary to use the plt.show() command

## Different Types of Plots

Line Plot

A line plot is a type of plot used to visualize the relationship between two continuous variables. It is a plot that displays data points as connected line segments. Line plots are commonly used to visualize trends and patterns in data, especially when the data is continuous and has a large number of observations.

In a line plot, the x-axis represents one variable and the y-axis represents another variable. The data points are connected by lines to create a plot that shows how the values of the two variables change with respect to each other. Line plots are often used to visualize time-series data, where the x-axis represents time and the y-axis represents a measurement.

In Matplotlib, the plot function is used to create a line plot. You can specify the x and y values of the data points, along with other optional arguments such as color, line style, and markers . You can also add labels, titles, and legends to your line plot to provide context and clarity.



```python
plt.style.use('seaborn-v0_8-dark')
x=np.linspace(0,4*np.pi,100)
y=np.sin(x)
cos_y=np.cos(x)
plt.plot(x,y)
plt.title('sine wave')
plt.show()
```

## 2D Histograms

plt.hist2d is a function in matplotlib that generates a 2D histogram. A 2D histogram is a representation of the distribution of data over two variables. In contrast to a regular histogram which shows the distribution of a single variable, a 2D histogram shows the relationship between two variables. The function creates a heatmap that shows the density of data points in a 2D space.

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.hist2d(x,y,bins=50,cmap="viridis")
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```

```python
plt.hist2d(x,y,bins=50,cmap="Blues")
plt.ylabel('Y')
plt.xlabel('X')
plt.colorbar()
plt.show()
```

## Pie Charts

A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion.

```python
cars=['honda','maruti','kia','bmw','audi','mercedes']
data=[200,350,50,25,250,10]
fig=plt.figure(figsize=(5,5))
plt.pie(data,labels=cars)
plt.show()
```

`Operations on pie chart`

```python
cars=['honda','maruti','kia','bmw','audi','mercedes']
data=[200,350,50,25,250,10]
explode=[0,0,0,0,0,0.2]
fig=plt.figure(figsize=(5,5))
plt.pie(data,labels=cars,explode=explode)
plt.show()
```

Explode Operation

```python
cars=['honda','maruti','kia','bmw','audi','mercedes']
data=[200,350,50,25,250,10]
explode=[0,0,0,0,0,0.2]
fig=plt.figure(figsize=(5,5))
plt.pie(data,labels=cars,explode=explode,shadow=True) # - Separates slices slightly for emphasis (e.g., [0.1, 0, 0, 0] would pull the first slice outward).
plt.show() 
```

Adding percentage

```python
cars=['honda','maruti','kia','bmw','audi','mercedes']
data=[200,350,50,25,250,10]
explode=[0,0,0,0,0,0.2]
fig=plt.figure(figsize=(5,5))
plt.pie(data,labels=cars,explode=explode,autopct='%2.2f%%') #autopct displays percentages inside slices with two decimal places.
plt.show()
```

```python
cars=['honda','maruti','kia','bmw','audi','mercedes']
data=[200,350,50,25,250,10]
explode=[0,0,0,0,0,0.2]
fig=plt.figure(figsize=(7,7))
plt.pie(data,labels=cars,explode=explode,autopct='%2.2f%%',wedgeprops={'edgecolor':'black','linewidth': 2,'antialiased':True})
# - edgecolor='black': Adds a black border around slices.
# - linewidth=2: Sets the border thickness.
# - antialiased=True: Smooths the edges for better visual quality.
plt.show()
```

##  Multiple Subplots

Multiple subplots refer to dividing a figure into multiple subplots or axes, each of which can contain a separate plot. This allows for multiple plots to be displayed in a single figure. In matplotlib, you can create multiple subplots using the subplot function.

Index Based

```python
x=np.linspace(0,14,50)
fig,ax=plt.subplots(2,2)
y=np.sin(x)
ax[0,0].plot(x,y)
ax[0,1].set_title('Sin Wave')
y=np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cos Wave')
y=np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan Wave')
y = np.random.randn(1000)
ax[1,1].hist(y,bins = 40)
ax[1,1].set_title('Random Distribution')
plt.tight_layout()
plt.show()
```

`With Loops - Example 1`

```python
plt.subplots_adjust(hspace=1,wspace=1)
for i in range(1,8):
    plt.subplot(3,3,i)
    plt.text(0.5,0.5,str((2,3,i)),ha="center")
```

`With Loops - Example 2`

```python
x = np.linspace(0,10,50)
y = np.sin(x)


fig , (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (18,6))
ax1.plot(x,y)
ax2.plot(x,y**2)
ax3.plot(x,y**3)

ax1.set_title('Sin Wave')
ax2.set_title('Sin Wave - Square')
ax3.set_title('Sin Wave - Cube')

plt.tight_layout()

plt.show()
```
