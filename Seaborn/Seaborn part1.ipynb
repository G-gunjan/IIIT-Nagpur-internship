## Scatter plot
A scatter plot can be used to demonstrate relationship between two variables x and y. A simple scatter plot can be drawn as follows:-

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='time')
```

```python
# scatter plot -> axes level function
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex',style='time',size='size')
```

```python
# rel plot->figure level->square shape
sns.relplot(x='total_bill',y='tip',data=tips,hue='sex',style='time',size='size')
```

```python
# line plot
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
temp_df=gap[gap['country'].isin(['India','Afghanistan','China'])]
temp_df
```

```python
sns.relplot(x='year',y='lifeExp',data=temp_df,hue='country',kind='line')
```

```python
# facet plot->figure level function->work with relplot
sns.relplot(data=tips,x='total_bill',y='tip',col='sex',row='day') # separated by sex across columns and day across rows
```

```python
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3) # separated by years across columns and warps 3 graphs in a row
```

##  Histogram


Histograms represent the data distribution by forming bins along the range of the data and then drawing bars to show 
the number of observations that fall in each bin.

We can use the same **histplot()** function to plot a histogram as follows:-


```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
sns.histplot(tips['total_bill'])
```

```python
sns.histplot(tips['total_bill'],kde=False) #kde=False removes kdeplot from graph
```

```python
# Plot a histogram of the total_bill column with 2 bins 
sns.histplot(tips['total_bill'], bins=2)
```

```python
# Plot a histogram of total_bill using custom bin edges to visualize distribution in specified intervals
sns.histplot(tips['total_bill'], bins=[5, 10, 20, 30, 40, 55])
```

```python
# Itâ€™s also possible to visualize the distribution of a categorical variable using the logic of a histogram. 
# Discrete bins are automatically set for categorical variables

# countplot
sns.histplot(tips['day'])
```

```python
sns.histplot(data=tips,x='total_bill',hue='sex') #plots histogram for each sex
```

```python
sns.histplot(data=tips,x='total_bill',hue='sex',element='step') #removes edges
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', kind='hist',element='step',hue='sex')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=tips, x='tip',element='step',col='sex')
```

## Seaborn â€“ Kernel Density Estimates


Kernel Density Estimation (KDE) is a way to estimate the probability density function of a continuous random variable. 
It is used for non-parametric analysis. 

Setting the hist parameter to false in distplot() function will yield the kernel density estimation plot.


```python
sns.distplot(tips['total_bill'], hist=False) 

plt.show()
```

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tips,x='total_bill')
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True,height=5,aspect=2) #width becomes 10 inches
```

```python
#Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data=tips, x='total_bill', y='tip')
```

##  Heat map


Seaborn heatmap() function plot rectangular data as a color-encoded matrix.
