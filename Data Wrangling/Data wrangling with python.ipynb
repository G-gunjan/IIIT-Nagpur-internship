merge and join function

```python
# Create two DataFrames with a common column 'A' and different values in column 'B'
var1 = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [12, 13, 14, 15]})
var2 = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [16, 17, 18, 19]})

# Merge the two DataFrames on column 'A'
pd.merge(var1, var2, on='A')
```

```python
var1=pd.DataFrame({'A':[1,2,3,4],'B':[12,13,14,15]})
var2=pd.DataFrame({'A':[1,2,3,4],'B':[16,17,18,19]})
pd.merge(var1,var2,on='A',how='left')
#It merges `var1` and `var2` on column `'A'`, keeping all rows from `var1` and adding matching `'B'` values from `var2` as a new column.
```

```python
var1=pd.DataFrame({'A':[1,2,3,4],'B':[12,13,14,15]})
var2=pd.DataFrame({'A':[1,2,3,4],'B':[16,17,18,19]})
pd.merge(var1,var2,on='A',how='right')
#It merges `var1` and `var2` on column `'A'`, keeping all rows from `var2` and adding matching `'B'` values from `var1`.So, the `right` join ensures all values from `var2` (right DataFrame) are preserved, and matches from `var1` are added alongside.
```

```python
var1=pd.DataFrame({'A':[1,2,3,4],'B':[12,13,14,15]})
var2=pd.DataFrame({'A':[1,2,3,5],'B':[16,17,18,19]})
# Perform an outer merge on 'A' to include all rows from both DataFrames, matching rows where 'A' values align
pd.merge(var1,var2,how='outer')
```

```python
var1=pd.DataFrame({'A':[1,2,3,4],'B':[12,13,14,15]})
var2=pd.DataFrame({'A':[1,2,3,5],'B':[16,17,18,19]})
pd.merge(var1,var2,left_index=True,right_index=True)
```

join and append

```python
df1 = pd.DataFrame({"A": [11, 12, 23, 34], "B": [22, 23, 24, 25]}, index=["a", "b", "c", "d"])
df2 = pd.DataFrame({"C": [21, 67], "D": [25, 23]}, index=["a", "b"])

# Join df2 to df1 based on the index, aligning rows where the index matches
df1.join(df2)
```

```python
df1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
df2=pd.DataFrame({"C":[10,20],"D":[11,22]},index=["a","b"])
df1.join(df2,how="outer")#outer performs union
```

```python
df1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
df2=pd.DataFrame({"C":[10,20],"D":[11,22]},index=["a","b"])
df1.join(df2,how="inner")#inner performs intersection
```

```python
df1=pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
df2=pd.DataFrame({"C":[10,20],"B":[11,22]},index=["a","b"])
df1.join(df2,how="outer",lsuffix="_12",rsuffix="_123")#to remove conflict
```

## Melt Function

```python
df=pd.DataFrame({'days':[1,2,3,4,5,6],'eng':[12,13,15,13,14,17],'maths':[10,11,12,13,14,15]})
df
```

```python
pd.melt(df) 
#The pd.melt() function un-pivots from dataFrame from wide format to long format.
```

```python
pd.melt(df,id_vars="days",var_name="subject",value_name="marks") #days stay as a fixed column subject and marks remain unpivoted
```

```python
df2=pd.DataFrame({"days":[1,2,3,4,5,6],"st_name":["a","b","a","b","a","b"],"eng":[10,12,13,14,15,16],"maths":[12,13,14,15,16,17]})
df2
```

```python
df2.pivot(index="st_name",columns="eng") #st_name will become new row label and eng will become new column label
```

```python
df2.pivot_table(index="st_name",columns="days",aggfunc="mean",margins=True) #adds new col All where avg is taken of all rows
```
